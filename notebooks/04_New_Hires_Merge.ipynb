```python
from pyspark.sql import Row
import random

new_hires = []

for emp_id in range(2069, 2089):
    new_hires.append(
        Row(
            Age=random.randint(22, 35),
            Attrition="No",
            BusinessTravel="Travel_Rarely",
            DailyRate=random.randint(500, 1500),
            Department=random.choice([
                "Sales",
                "Research & Development",
                "Human Resources"
            ]),
            DistanceFromHome=random.randint(1, 20),
            Education=random.randint(2, 5),
            EducationField="Life Sciences",
            EmployeeCount=1,
            EmployeeNumber=emp_id,
            EnvironmentSatisfaction=random.randint(2, 4),
            Gender=random.choice(["Male", "Female"]),
            HourlyRate=random.randint(30, 100),
            JobInvolvement=random.randint(2, 4),
            JobLevel=1,
            JobRole=random.choice([
                "Sales Executive",
                "Research Scientist",
                "Laboratory Technician"
            ]),
            JobSatisfaction=random.randint(2, 4),
            MaritalStatus=random.choice([
                "Single",
                "Married"
            ]),
            MonthlyIncome=random.randint(4000, 9000),
            MonthlyRate=random.randint(10000, 25000),
            NumCompaniesWorked=random.randint(0, 3),
            Over18="Y",
            OverTime="No",
            PercentSalaryHike=random.randint(10, 20),
            PerformanceRating=3,
            RelationshipSatisfaction=random.randint(2, 4),
            StandardHours=80,
            StockOptionLevel=random.randint(0, 2),
            TotalWorkingYears=random.randint(1, 8),
            TrainingTimesLastYear=random.randint(1, 4),
            WorkLifeBalance=random.randint(2, 4),
            YearsAtCompany=0,
            YearsInCurrentRole=0,
            YearsSinceLastPromotion=0,
            YearsWithCurrManager=0
        )
    )

new_hires_df = spark.createDataFrame(new_hires)

display(new_hires_df)
```
