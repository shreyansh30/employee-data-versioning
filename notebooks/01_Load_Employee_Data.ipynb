```python
%sql
CREATE DATABASE IF NOT EXISTS employee_db;
```

```python
df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("employee_db.employee_versioned")
```

```python
%sql
SELECT COUNT(*) 
FROM employee_db.employee_versioned;
```

```python
%sql
DESCRIBE DETAIL employee_db.employee_versioned;
```
